```python
categorical_map = {'not fire   ':0,'fire   ':1,'fire':1,'fire ':1,'not fire':0,'not fire ':0,'Classes  ':1,'not fire     ':0, np.nan :0,'not fire    ':0}
```

```python
df['Classes'] = df['Classes  '].map(categorical_map)
```

```python
df.drop(['Classes  '],axis = 1,inplace = True)
```

```python
df.columns = df.columns.str.strip()
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Rain']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Rain']].apply(pd.to_numeric, errors='coerce')
```

```python
df.fillna(0, inplace=True)  # Or use another strategy for handling NaNs
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Rain']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Rain']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df[['FFMC','DMC','DC','ISI','BUI','FWI']] = df[['FFMC','DMC','DC','ISI','BUI','FWI']].apply(pd.to_numeric, errors='coerce')
```

```python
df.fillna(0,inplace = True)
```

```python
df[['FFMC','DMC','DC','ISI','BUI','FWI']]  = df[['FFMC','DMC','DC','ISI','BUI','FWI']] .astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('cleaned_Algerian_Forest',index = False)
```

```python
df_copy  = df.drop(['day','month','year'],axis=1)
```

```python
df_copy
```

```python
df_copy['Clasees'] = np.where(df_copy['Classes']=='not fire',0,1)
```

```python
percentage = df_copy['Classes'].value_counts(normalize = True)*100
```

```python
# plotting piechart
classlabel = ['Fire' , 'Not FIre']
plt.figure(figsize = (10,7))
plt.pie(percentage,labels = classlabel)
plt.title('Pie chart for the classes')
plt.show()
```
